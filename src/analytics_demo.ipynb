This notebook demonstrates how learning analytics can visualize engagement and collaboration data generated by the AI prototype.
All data below are synthetic and used only for demonstration.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Synthetic collaboration data
data = {
    'Student': ['Alice', 'Ben', 'Cleo', 'David'],
    'Projects_Completed': [3, 5, 2, 4],
    'Collaboration_Score': [85, 92, 78, 88],
    'Reflection_Score': [80, 89, 75, 83]
}

df = pd.DataFrame(data)
df
```

```python
# Bar chart: Project completion and collaboration scores
plt.figure(figsize=(7,4))
plt.bar(df['Student'], df['Collaboration_Score'], color='skyblue', label='Collaboration')
plt.bar(df['Student'], df['Reflection_Score'], color='lightgreen', alpha=0.6, label='Reflection')
plt.title('Collaboration and Reflection Scores by Student')
plt.xlabel('Student')
plt.ylabel('Score')
plt.legend()
plt.tight_layout()
plt.show()
```

### Interpretation
This visualization represents how learning analytics can summarize collaboration patterns in project-based courses.
- Higher scores indicate more active group participation.
- Differences between collaboration and reflection scores may indicate gaps between participation and learning depth.

In a real deployment, these metrics could be updated dynamically from student interactions and displayed in an instructor dashboard.
